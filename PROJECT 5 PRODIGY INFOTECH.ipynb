```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\tamil\Downloads\Prodigy_Dataset_05.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df['Timestamp'] = pd.to_datetime(df['Time'])
df['Hour'] = df['Timestamp'].dt.hour
df['DayOfWeek'] = df['Timestamp'].dt.day_name()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Hour'], bins=24, kde=True)
plt.title("Accident Distribution by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(y=df['Weather_conditions'], order=df['Weather_conditions'].value_counts().index)
plt.title("Accident Distribution by Weather Condition")
plt.xlabel("Number of Accidents")
plt.ylabel("Weather Condition")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(y=df['Road_surface_conditions'], order=df['Road_surface_conditions'].value_counts().index)
plt.title("Accident Distribution by Road Condition")
plt.xlabel("Number of Accidents")
plt.ylabel("Road Condition")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(y=df['Accident_severity'], order=df['Accident_severity'].value_counts().index)
plt.title("Accident Severity Distribution")
plt.xlabel("Number of Accidents")
plt.ylabel("Accident Severity")
plt.show()
```
