```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from textblob import TextBlob
```

```python
df = pd.read_csv(r"C:\Users\tamil\Music\twitter_training.csv",header=None,names=['ID', 'Topic', 'Sentiment_Label', 'Text'])  
df.head()
```

```python
def get_sentiment(text):
    return TextBlob(str(text)).sentiment.polarity
```

```python
df['Sentiment_Score'] = df['Text'].apply(get_sentiment)
df['Predicted_Sentiment'] = df['Sentiment_Score'].apply(lambda x: 'Positive' if x > 0 else ('Negative' if x < 0 else 'Neutral'))
```

```python
sns.countplot(x=df['Predicted_Sentiment'], palette="coolwarm")
plt.title("Sentiment Distribution (Predicted)")
plt.show()
```

```python
sns.countplot(x=df['Sentiment_Label'], palette="coolwarm")
plt.title("Sentiment Distribution (Original)")
plt.show()
```

```python
text = " ".join(str(review) for review in df['Text'])
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)
```

```python

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
