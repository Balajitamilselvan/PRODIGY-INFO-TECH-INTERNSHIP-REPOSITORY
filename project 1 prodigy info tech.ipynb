```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(r"C:\Users\tamil\Music\pit_1.csv",skiprows=4)
df
```

```python
df.isnull().sum()
```

```python
df_1=df[["Country Name","Country Code","2020"]]
df_1.head()
```

```python
df_1.isnull().sum()
```

```python
selected_countries = ["India", "United States", "China", "Brazil", "Germany"]
df_filtered = df_1[df_1["Country Name"].isin(selected_countries)]
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x="Country Name", y="2020", data=df_filtered)
plt.xlabel("Country")
plt.ylabel("Population in 2020")
plt.title("Population of Selected Countries in 2020")
plt.xticks(rotation=45)
plt.show()
```

```python
df = df.drop(columns=["Indicator Name"], errors="ignore")
```

```python
year_columns = [col for col in df.columns if col.isdigit()] 
```

```python
df_melted = df.melt(id_vars=["Country Name", "Country Code"], 
                    value_vars=year_columns, 
                    var_name="Year", value_name="Population")
```

```python
df_melted
```

```python
df_melted["Year"] = pd.to_numeric(df_melted["Year"], errors="coerce")  
df_melted.dropna(inplace=True)  
df_india = df_melted[df_melted["Country Name"] == "India"]
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df_india["Population"], bins=20, kde=True)
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.title("Population Distribution of India Over Time")
plt.show()
```
